```python
for indicator in list_indicators:
    if(indicator == "Aid received"):
        continue
#for indicator in ["Completition rate female"]:
    #list_countries = global_df.loc[global_df["indicator"] == indicator,["name"]].unique()
    countries_with_no_data = []

    for country in list_countries:
    #for country in ["Iraq"]:
        min_year = global_df.loc[global_df["indicator"] == indicator, ["year"]].min().values[0]
        max_year =  global_df.loc[global_df["indicator"] == indicator, ["year"]].max().values[0]
        
        year_df_not_null = global_df.loc[~(global_df["value"].isnull()) 
                  & (global_df["indicator"] == indicator)
                  & (global_df["name"] == country), ["year"]].sort_values(by="year")
        
        list_year_not_null = [item for sublist in year_df_not_null.values for item in sublist]

        if(len(list_year_not_null) == 0):
            countries_with_no_data.append(country)
            continue
        
        # Mise à jour des premières valeurs si vides avec la première non vide
        #print(min_year)
        #print(max_year)
        #print(list_year_not_null)
        if(list_year_not_null[0] > min_year):
            #print("maj first values")
            value_first_indicator = float(global_df.loc[(global_df["indicator"] == indicator) 
                                      & (global_df["name"] == country) 
                                      & (global_df["year"] == list_year_not_null[0]), ["value"]].value)
            #print(value_first_indicator)
            for year in np.arange(min_year, list_year_not_null[0]):
                global_df.loc[(global_df["indicator"] == indicator) 
                                      & (global_df["name"] == country) 
                                      & (global_df["year"] == year), ["value"]] = value_first_indicator
            
   
        # Mise à jour des dernières valeurs si vides avec la dernière non vide
        if(list_year_not_null[-1] < max_year):
            #print("maj last values")
            value_last_indicator = float(global_df.loc[(global_df["indicator"] == indicator) 
                                      & (global_df["name"] == country) 
                                      & (global_df["year"] == list_year_not_null[-1]), ["value"]].value)
            #print(value_last_indicator)
            for year in np.arange(list_year_not_null[-1], max_year + 1):
                global_df.loc[(global_df["indicator"] == indicator) 
                                      & (global_df["name"] == country) 
                                      & (global_df["year"] == year), ["value"]] = value_last_indicator
    
    
        # Mise à jour des valeurs intermédiaires
        previous_year = list_year_not_null[0]
        nb_year_null = 0
        for current_year in list_year_not_null:
            if(current_year - previous_year > 1):
                nb_year_null = current_year - previous_year - 1
                value_previous_year = float(global_df.loc[(global_df["indicator"] == indicator) 
                                                     & (global_df["name"] == country) 
                                                     & (global_df["year"] == previous_year), ["value"]].value)
                value_current_year = float(global_df.loc[(global_df["indicator"] == indicator) 
                                                     & (global_df["name"] == country) 
                                                     & (global_df["year"] == current_year), ["value"]].value)
                step_value = (value_current_year - value_previous_year) / (nb_year_null + 1)

                for i, year in enumerate(np.arange(previous_year + 1, current_year)):
                    global_df.loc[(global_df["indicator"] == indicator) 
                                  & (global_df["name"] == country) 
                                  & (global_df["year"] == year), ["value"]] = value_previous_year + (i + 1) * step_value
            
            previous_year = current_year
            
    print(indicator)
    #if (len(countries_with_no_data) > 0):
    #   print("Pas de valeur pour l'indicteur : {}, pour les pays : {}".format(indicator, countries_with_no_data))
```

## Mise à l'échelle - min/max par année

```python
file_path = "../static/data/scaled_indicators.csv"
global_df_scaled = pd.read_csv(file_path, sep=";")
```

```python
def scale_value(value, min_val, max_val):
    scale_val = 100 * (value - min_val) / (max_val - min_val) + 25
    return scale_val
```

```python
global_df_scaled = global_df_no_nan.copy()
global_df_scaled["value_scaled"] = global_df_scaled["value"]
```

```python
del global_df_scaled["Unnamed: 0"]
del global_df_scaled["Unnamed: 0.1"]
```

```python
indicator_to_rescale = global_df_scaled["indicator"].unique()
```

```python
#for indicator in list_indicators:
for indicator in indicator_to_rescale:
    df_one_indicator = global_df_scaled[global_df_scaled["indicator"] == indicator]
    
    min_year = global_df_scaled.loc[global_df_scaled["indicator"] == indicator, ["year"]].min().values[0]
    max_year =  global_df_scaled.loc[global_df_scaled["indicator"] == indicator, ["year"]].max().values[0]
    
    for year in np.arange(min_year, max_year + 1):
        min_value = global_df_scaled.loc[(global_df_scaled["year"] == year) 
                                         &(global_df_scaled["indicator"] == indicator),
                                         ["value"]].min().values[0]
        
        max_value = global_df_scaled.loc[(global_df_scaled["year"] == year) 
                                         &(global_df_scaled["indicator"] == indicator),
                                         ["value"]].max().values[0]
        
        global_df_scaled.loc[(global_df_scaled["indicator"] == indicator)
                             & (global_df_scaled["year"] == year),
                             ["value_scaled"]] = global_df_scaled[(global_df_scaled["indicator"] == indicator)
                             & (global_df_scaled["year"] == year)].apply(axis=1, func= lambda x: scale_value(x["value"], min_value, max_value))
        
```
